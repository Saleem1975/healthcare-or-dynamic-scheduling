# CNN-Based Dynamic OR Scheduling Demo

This notebook uses the sample schedule-comparison data prepared from the published paper to reproduce the main operational-impact calculations:

- Total patient waiting time under traditional static scheduling
- Total patient waiting time under CNN-based dynamic scheduling
- Percentage reduction in patient waiting time
- Operating room idle time reduction
- Simple visualization of the before/after impact

Raw laparoscopic videos are not included in this repository for privacy and clinical data-sharing reasons.


```
import pandas as pd
import matplotlib.pyplot as plt

data_path = "../data/sample_schedule_data.csv"
df = pd.read_csv(data_path)
df
```

## Calculate total waiting time and OR idle time

```
traditional_wait = df["traditional_waiting_time_min"].sum()
cnn_wait = df["cnn_waiting_time_min"].sum()
waiting_reduction = (traditional_wait - cnn_wait) / traditional_wait * 100

traditional_idle = df["traditional_or_idle_time_min"].sum()
cnn_idle = df["cnn_or_idle_time_min"].sum()

print(f"Traditional total patient waiting time: {traditional_wait} minutes")
print(f"CNN-based total patient waiting time: {cnn_wait} minutes")
print(f"Patient waiting time reduction: {waiting_reduction:.1f}%")
print(f"Traditional OR idle time: {traditional_idle} minutes")
print(f"CNN-based OR idle time: {cnn_idle} minutes")
```

## Waiting time comparison by operation

```
plt.figure(figsize=(10, 6))
plt.plot(df["operation_number"], df["traditional_waiting_time_min"], marker="o", linewidth=2, label="Traditional static scheduling")
plt.plot(df["operation_number"], df["cnn_waiting_time_min"], marker="o", linewidth=2, label="CNN-based dynamic scheduling")
plt.title("Patient Waiting Time by Operation")
plt.xlabel("Operation number")
plt.ylabel("Waiting time (minutes)")
plt.xticks(df["operation_number"])
plt.grid(True, linestyle="--", alpha=0.4)
plt.legend()
plt.tight_layout()
plt.show()
```

## Summary impact chart

```
impact = pd.DataFrame({
    "Metric": ["Patient waiting time", "OR idle time"],
    "Traditional": [traditional_wait, traditional_idle],
    "CNN-based": [cnn_wait, cnn_idle]
})

ax = impact.set_index("Metric").plot(kind="bar", figsize=(9, 6), rot=0)
ax.set_title("Operational Impact of CNN-Based Dynamic Scheduling")
ax.set_ylabel("Minutes")
ax.grid(axis="y", linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()

impact
```

## Interpretation

In this schedule demonstration, CNN-based dynamic scheduling reduces patient waiting time from **103 minutes** to **13 minutes**, an **87.3% reduction**. It also reduces OR idle time from **28 minutes** to **0 minutes**.

This supports the key operational message of the project: real-time phase detection can help hospitals update schedules dynamically and improve operating room efficiency.
